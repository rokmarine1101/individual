```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
import seaborn as sns
import numpy as np
import statsmodels.api as sm
from sklearn.datasets import make_regression
from patsy import dmatrix
from sklearn.linear_model import LinearRegression
import statsmodels.stats.api as sms
import numpy as np
import re
```

```python
df = pd.read_csv("lotto.csv")
```

```python
def even_or_odd(x):
    if x % 2 != 0: return 'odd'
    else: return 'even'
```

```python
first_nums = df["1st"].apply(even_or_odd)
second_nums = df["2nd"].apply(even_or_odd)
third_nums = df["3rd"].apply(even_or_odd)
fourth_nums = df["4th"].apply(even_or_odd)
fifth_nums = df["5th"].apply(even_or_odd)
sixth_nums = df["6th"].apply(even_or_odd)
```

```python
sns.countplot(first_nums)
plt.show()
sns.countplot(second_nums)
plt.show()
sns.countplot(third_nums)
plt.show()
sns.countplot(fourth_nums)
plt.show()
sns.countplot(fifth_nums)
plt.show()
sns.countplot(sixth_nums)
plt.show()
```
